```python
import json
from matplotlib import pyplot as plt
```

```python
def load_json(path):
    with open(path, 'r') as f:
        return json.load(f)
```

```python
nn_results = load_json('./results/nn_results.json')
```

```python
def plot_nn_results(results):
    plt.xlabel('Episodes')
    plt.ylabel('Reward')
    plt.ylim((0, 220))
    plt.title('Neural Network Results')
    for result in results:
        name = result['algo_name']
        records = result['records']
        x = [i for i in range(len(records))]
        plt.plot(x, records, label=name)
    plt.legend(loc='upper right')
    plt.savefig('./results/image/nn_result.png', dpi=150)
    plt.show()
```

```python
plot_nn_results(nn_results)
```

```python
def plot_genetic_results(results):
    plt.xlabel('Generations')
    plt.ylabel('Best Reward')
    plt.ylim((0, 220))
    plt.title('Genetic Algorithm Results')
    for result in results:
        name = f'pop = {result["num_pops"]}'
        records = result['records']
        x = [i for i in range(len(records))]
        plt.plot(x, records, label=name)
    plt.legend(loc='upper right')
    plt.savefig('./results/image/genetic_result.png', dpi=150)
    plt.show()
```
